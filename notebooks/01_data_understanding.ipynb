```python
import pandas as pd
# Load the UCI Student Performance mathematics dataset
# The CSV file uses semicolon ; as the separator
# '../' moves up to main folder, then we navigate into data/raw/
file_path = "../data/raw/student-mat.csv"

df = pd.read_csv(file_path, sep=";")
print("Dataset shape:", df.shape)
df.head()
```

```python
summary = pd.DataFrame({
"variable": df.columns,
"dtype": df.dtypes.astype(str).values,
"n_unique": df.nunique().values,
"n_missing": df.isna().sum().values,
})
display(summary)
```

# S9: Data Inspection
In this section, we inspect the dataset structure before cleaning and modeling.
We check the dataset shape, column data types, missing-value information, and
summary statistics.

```python
# S9: Standard Inspection Trio
print("1. Dataset shape:")
print(df.shape)
print("\n2. Dataset information:")
df.info()
print("\n3. Summary statistics for all variables:")
display(df.describe(include="all").T)
```

## S9 Inspection Summary Note
The dataset inspection shows the number of rows and columns, the variable names,
the data types, and the basic summary statistics. The `df.info()` output helps
identify whether variables are numeric or categorical and whether any columns
contain missing values.
From the `df.describe(include="all").T` output, three important observations are:
1. The dataset contains 395 rows and 33 columns.
2. Some variables are categorical and will need encoding before machine learning.
3. Some numeric variables should be checked for unusual minimum values, maximum
values, or wide ranges.
The next cleaning step is to check missing values, separate numeric and categorical
columns, encode categorical variables, and confirm the target variable for
